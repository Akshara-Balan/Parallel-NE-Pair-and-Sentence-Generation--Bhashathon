```python
!python -m spacy download en_core_web_trf
!pip install datasets pandas spacy transformers
!pip install -q torch torchvision torchaudio  # Ensure PyTorch is installed

import pandas as pd
import spacy
from transformers import AutoModelForTokenClassification, AutoTokenizer, pipeline
from datasets import load_dataset
```

```python
# Load the dataset
df = pd.read_csv("/content/output.csv")

# Check first few rows
print(df.head())
```

```python
# Load English NER model (spaCy)
nlp_en = spacy.load("en_core_web_trf")
from huggingface_hub import login
from google.colab import userdata
userdata.get('HFRead')


# Load Malayalam NER model (ai4bharat/indic-ner)
model_name = "ai4bharat/indicNer"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForTokenClassification.from_pretrained(model_name)
ner_pipeline = pipeline("ner", model=model, tokenizer=tokenizer)
```

```python
def extract_english_ner(text):
    if pd.isna(text):
        return []
    doc = nlp_en(text)
    return [ent.text for ent in doc.ents]

def extract_malayalam_ner(text):
    if pd.isna(text):
        return []
    results = ner_pipeline(text)
    return [r["word"] for r in results]

# Extract Named Entities
df["English_NEs"] = df["src"].apply(extract_english_ner)
df["Malayalam_NEs"] = df["tgt"].apply(extract_malayalam_ner)

# Save extracted NE dataset
df.to_csv("en_ml_dataset_with_NEs.csv", index=False)

print("✅ Named Entity Extraction Completed! Dataset saved as 'en_ml_dataset_with_NEs.csv'.")
```

```python
import pandas as pd
from fuzzywuzzy import fuzz
from rapidfuzz import process

# Load dataset
df = pd.read_csv("en_ml_dataset_with_NEs.csv")

# 1️⃣ **Remove Duplicates & Near-Duplicates**
def remove_duplicates(df, threshold=90):
    unique_pairs = {}
    filtered_rows = []

    for index, row in df.iterrows():
        src, tgt = row["src"], row["tgt"]
        key = (src.lower(), tgt.lower())

        if key not in unique_pairs:
            unique_pairs[key] = index
            filtered_rows.append(row)
        else:
            # If similarity is high, remove duplicate
            sim_score = fuzz.ratio(src, tgt)
            if sim_score < threshold:
                filtered_rows.append(row)

    return pd.DataFrame(filtered_rows)

df = remove_duplicates(df)

# 2️⃣ **Filter by Length Ratio (Too Short or Too Long Translations)**
def filter_length_ratio(df, min_ratio=0.5, max_ratio=2.0):
    def valid_length_ratio(src, tgt):
        ratio = len(src.split()) / max(1, len(tgt.split()))
        return min_ratio <= ratio <= max_ratio

    return df[df.apply(lambda row: valid_length_ratio(row["src"], row["tgt"]), axis=1)]

df = filter_length_ratio(df)

# 3️⃣ **Handle Named Entity Mismatches (NE Missing in One Sentence)**
def filter_named_entity_mismatches(df):
    def has_mismatch(row):
        en_entities = set(eval(row["English_NEs"])) if isinstance(row["English_NEs"], str) else set()
        ml_entities = set(eval(row["Malayalam_NEs"])) if isinstance(row["Malayalam_NEs"], str) else set()
        return bool(en_entities - ml_entities or ml_entities - en_entities)

    return df[~df.apply(has_mismatch, axis=1)]

df = filter_named_entity_mismatches(df)

# Save cleaned dataset
df.to_csv("filtered_parallel_sentences.csv", index=False)
print("✅ Parallel sentence filtering completed! Check 'filtered_parallel_sentences.csv'.")
```

```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

# Load a paraphrasing model (T5-based for English)
paraphrase_model = AutoModelForSeq2SeqLM.from_pretrained("Vamsi/T5_Paraphrase_Paws")
paraphrase_tokenizer = AutoTokenizer.from_pretrained("Vamsi/T5_Paraphrase_Paws")

# Function to generate paraphrases
def generate_paraphrase(text, max_length=50):
    input_text = f"paraphrase: {text} </s>"
    encoding = paraphrase_tokenizer.encode_plus(input_text, return_tensors="pt", max_length=512, truncation=True)
    outputs = paraphrase_model.generate(**encoding, max_length=max_length, num_return_sequences=1)
    return paraphrase_tokenizer.decode(outputs[0], skip_special_tokens=True)

# Test the function
print("Original:", "The government is planning new policies.")
print("Paraphrase:", generate_paraphrase("The government is planning new policies."))
```

```python
import wikipediaapi

# Define a descriptive user agent string
user_agent = "MyWikipediaApp/1.0 (https://example.com; contact@example.com)"

# Create the Wikipedia object, passing the user_agent
wiki_wiki = wikipediaapi.Wikipedia(language='en',  # Specify the language
                                   extract_format=wikipediaapi.ExtractFormat.WIKI,
                                   user_agent=user_agent)  # Pass the user agent string here

def get_wikipedia_summary(entity):
    try:
        page = wiki_wiki.page(entity)
        if page.exists():
            return page.summary.split(".")[0]  # Get first sentence of summary
    except:
        return None
    return None

# Test the function
print(get_wikipedia_summary("Narendra Modi"))
```

```python
import pandas as pd
import random

# Load augmented dataset
df = pd.read_csv("augmented_parallel_sentences.csv")

# Define entity replacement dictionary (Expand this!)
entity_variants = {
    "India": ["Bharat", "Republic of India"],
    "United States": ["USA", "United States of America"],
    "New Delhi": ["Delhi", "NCR"],
    "Narendra Modi": ["Modi", "Indian Prime Minister"],
    "Microsoft": ["MSFT", "Tech Giant Microsoft"]
}

def replace_named_entities(text):
    for entity, variants in entity_variants.items():
        if entity in text:
            text = text.replace(entity, random.choice(variants))
    return text

# Apply entity replacements
df["entity_augmented_src"] = df["paraphrased_src"].apply(replace_named_entities)

# Save dataset
df.to_csv("entity_augmented_parallel_sentences.csv", index=False)
print("✅ Named Entity Augmented dataset saved as 'entity_augmented_parallel_sentences.csv'.")
```

```python
df = pd.read_csv("entity_augmented_parallel_sentences.csv")

# Assuming 'tgt' is the original target column
df["entity_augmented_tgt"] = df["tgt"]  # Create the missing column

# Compute sentence lengths
df["src_length"] = df["entity_augmented_src"].apply(lambda x: len(str(x).split()))
df["tgt_length"] = df["entity_augmented_tgt"].apply(lambda x: len(str(x).split()))

# Remove sentence pairs where one sentence is much shorter
df = df[(df["src_length"] / df["tgt_length"]).between(0.5, 2.0)]

print(f"✅ Removed length mismatched sentences. Remaining: {len(df)} pairs.")
```

```python
def compute_labse_similarity(src, tgt):
    embeddings = labse_model.encode([src, tgt])
    return np.dot(embeddings[0], embeddings[1]) / (np.linalg.norm(embeddings[0]) * np.linalg.norm(embeddings[1]))

df["translation_quality"] = df.apply(lambda row: compute_labse_similarity(row["entity_augmented_src"], row["entity_augmented_tgt"]), axis=1)

# Keep only high-quality translations
df = df[df["translation_quality"] > 0.7]  # Adjust threshold as needed

print(f"✅ Filtered low-quality translations. Remaining: {len(df)} pairs.")
```
