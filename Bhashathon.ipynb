```python
import pandas as pd
import spacy
import stanza

# Load dataset
df = pd.read_csv("en_ml_dataset.csv")

# Load NER models
nlp_en = spacy.load("en_core_web_trf")  # English model
stanza.download("ml")  # Download Malayalam model
nlp_ml = stanza.Pipeline(lang="ml", processors="tokenize,ner")  # Malayalam model

# Function to extract NEs
def extract_named_entities(text, lang="en"):
    if pd.isna(text):  # Handle missing values
        return []
    if lang == "en":
        doc = nlp_en(text)
        return [ent.text for ent in doc.ents]
    elif lang == "ml":
        doc = nlp_ml(text)
        return [ent.text for ent in doc.entities]

# Extract NEs for English and Malayalam
df["English_NEs"] = df["source"].apply(lambda x: extract_named_entities(x, "en"))
df["Malayalam_NEs"] = df["target"].apply(lambda x: extract_named_entities(x, "ml"))

# Save extracted NEs
df.to_csv("en_ml_dataset_with_NEs.csv", index=False)

print("Named Entity Extraction completed! Dataset saved as 'en_ml_dataset_with_NEs.csv'.")
```
