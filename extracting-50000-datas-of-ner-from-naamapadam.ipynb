Extracting 50000 datas with most number of named entities

```python
from datasets import load_dataset
import pandas as pd

# Load the Malayalam subset
dataset = load_dataset('ai4bharat/naamapadam', 'ml')

# Convert to a DataFrame
df = pd.DataFrame(dataset['train'])

# Function to count named entities (non-"O" tags)
def count_named_entities(tag_list):
    return sum(1 for tag in tag_list if tag != "O")

# Compute NE counts
df['NE_count'] = df['ner_tags'].apply(count_named_entities)

# Sort by NE count in descending order
df_sorted = df.sort_values(by='NE_count', ascending=False)

# Select top 50,000 rows
df_top_50k = df_sorted.head(50000)

# Save to CSV
df_top_50k.to_csv("top_50000_named_entities.csv", index=False)

print("Top 50,000 rows with max named entities saved as 'top_50000_named_entities.csv'")
```
